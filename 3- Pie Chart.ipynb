```python
from matplotlib import pyplot as plt

slice = [30,20,10]
Labels = ['Thirty','Twenty','Ten']
colors = ['#e5ae37','Blue','Green']
#Note: These labels and colors should be corrospondents to slices


plt.pie(slice,labels=Labels,colors=colors)
plt.title('Sample Pie Chart')

plt.show()


```

### wedgeprops property to add outlines in circle:

```python
from matplotlib import pyplot as plt

slice = [30,20,10]
Labels = ['Thirty','Twenty','Ten']
colors = ['purple','organge','Green']
#Note: These labels should be corrospondents to slices


plt.pie(slice,labels=Labels,wedgeprops={'edgecolor':'black'},)
plt.title('Sample Pie Chart')

plt.show()


```

### explode (to keep slice bit out from circle) and shadow property (to proper a bit of 3d effect)

```python
#explode property to provide emphasis or cut it out a bit form circle
from matplotlib import pyplot as plt

slice = [30,20,10]
Labels = ['Thirty','Twenty','Ten']
colors = ['purple','organge','Green']
explode = [0,0,.2]
#Note: These labels should be corrospondents to slices


plt.pie(slice,labels=Labels,wedgeprops={'edgecolor':'black'},explode=explode,shadow=True)
plt.title('Sample Pie Chart')

plt.show()

```

### Startangle(Rotation) and autopct='%1.1f%%' (To reveal percentage of each slice)

```python
from matplotlib import pyplot as plt

slice = [30,20,10]
Labels = ['Thirty','Twenty','Ten']
colors = ['purple','organge','Green']
explode = [0,0,.2]
#Note: These labels should be corrospondents to slices


plt.pie(slice,labels=Labels,wedgeprops={'edgecolor':'black'},explode=explode,shadow=True,startangle=345,autopct='%1.1f%%')
plt.title('Sample Pie Chart')

plt.show()
```
