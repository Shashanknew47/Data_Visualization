```python
#For Using matplot lib we will use mostly pyplot module. For using matplotlib in Jupyter notbooks we will use %matplotlib inline
import matplotlib.pyplot as plt
%matplotlib inline

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.show()
```

```python
'''
Source : Corey Shafer
You can add different formatting with format string. : For checking these properties explore this URL
(https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html   This reveals format strings reveal 3 things to change
fmt = '[marker][line][color]).
'''
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y,color='green',linewidth=2,marker='o')
plt.show()
```

```python
#Source : Corey Shaffer series
#To add different properties to the graph
x = [1,2,3,4,5]
y = [10,20,30,40,50]
y1 = [15,20,25,30,35]
plt.plot(x,y,label='yaxis',color='green',linewidth=2,linestyle='--',marker='o')
plt.plot(x,y1,label='y1axis',color='#0000ff',linewidth=3,linestyle='-.',marker='o') #you can put hex values for the color.

plt.xlabel('x-axis') 
plt.ylabel('y-axis')
plt.title('slaes_graph')
plt.grid(True) #square lines in background
#plt.legend(['yaxis','y1axis'])
plt.legend() #To add legend first we will add label in plt.plot after mentioning axises plt.plot(x,y,label='yaxis')
plt.show()
```

```python
# check above graph with some inbuilt styling.
#tight_layout will keep graph in concentrate optimum area.
x = [1,2,3,4,5]
y = [10,20,30,40,50]
y1 = [15,20,25,30,35]
plt.plot(x,y,marker='o',label='y-axis')
plt.plot(x,y1,marker='o',label='z-axis')
plt.title('Sales_graph')
plt.style.use('seaborn-pastel')
plt.legend()
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#Just like line charts we could make bar chart by plt.bar instead of plt.plt.

plt.style.use('dark_background')
x = [1,2,3,4,5]
y = [3,4,5,6,7]



plt.bar(x,y,label="end",color='b')



plt.xlabel('x-axis') 
plt.ylabel('y-axis')
plt.grid(True) #square lines in background
#plt.legend(['yaxis','y1axis'])
plt.legend() #To add legend first we will add label in plt.plot after mentioning axises plt.plot(x,y,label='yaxis')
plt.show()
```
