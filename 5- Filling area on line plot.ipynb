```python
#When you just want to highlight in red where python salries was lower than all other average developer salaries. and where
#it's high then it should be in green.
# To label each high and down trend we will label each condition in label.
# interpolate we will use to fill color properly otherwise it color might not be filled in every.



from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('C:/Users/Shashank/Desktop/Data Samples/Data_filling area lineplot.csv')
data

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages,dev_salaries,linestyle='--',label='All_Devs',color='#444444')
plt.plot(ages,py_salaries,label='Python')


plt.fill_between(ages,py_salaries,dev_salaries,alpha=.24,
                where=(py_salaries<dev_salaries),color='r',label='below_avg',interpolate=True)
plt.fill_between(ages,py_salaries,dev_salaries,alpha=.24,
                where=(py_salaries>dev_salaries),color='g',label='above_avg',interpolate=True)


plt.legend(loc=(.02,.65))

plt.title("Median Salary by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary")

plt.show()

```
