```python
#How to plot multiple bar graph in a single chart.
# We will numpy to overcome this problem
# For this we need to 1 bar either to the right or left another bar and
#have to set width of bars

from matplotlib import pyplot as plt
import numpy as np

plt.style.use('fivethirtyeight')

ages_x = range(25,36)
print(list(ages_x))

x_indexs = np.arange(len(ages_x))
width = .25


py_dev_y = [45372,48876,53850,57287,63016,65998,70003,7000,71496,75370,83640]
plt.plot(x_indexs,dev_y,color='g',label='Python')

js_dev_y = [37810,43515,46823,49293,53437,57373,62375,66674,68745,68746,74583]
plt.bar(x_indexs-width,js_dev_y,width=width,color='#e5ae38',label='Javascript')

dev_y  = [34896,42000,46752,49320,53200,56000,62316,64928,67312,68748,73752]
plt.bar(x_indexs,dev_y,width=width,color='r',label='All_devs')

plt.legend()

plt.title('Median Salary in USD by Age')

plt.xlabel('Ages')

plt.ylabel('Median Salary in USD')

#plt.tight_layout()


plt.show()
```

```python
#Now to set back x-axis back to the original label by using xticks method. 

from matplotlib import pyplot as plt
import numpy as np

#plt.style.use('fivethirtyeight')

ages_x = range(25,36)
print(list(ages_x))

x_indexs = np.arange(len(ages_x))
width = .25


py_dev_y = [45372,48876,53850,57287,63016,65998,70003,70008,71496,75370,83640]
plt.plot(x_indexs,py_dev_y,color='g',label='Python')

js_dev_y = [37810,43515,46823,49293,53437,57373,62375,66674,68745,68746,74583]
plt.bar(x_indexs-width,js_dev_y,width=width,color='#e5ae38',label='Javascript')

dev_y  = [34896,42000,46752,49320,53200,56000,62316,64928,67312,68748,73752]
plt.bar(x_indexs,dev_y,width=width,color='r',label='All_devs')

plt.legend()

plt.xticks(ticks=x_indexs,labels=ages_x)

plt.title('Median Salary in USD by Age')

plt.xlabel('Ages')

plt.ylabel('Median Salary in USD')

#plt.tight_layout()


plt.show()
```

```python
from matplotlib import pyplot as plt
import csv
from collections import Counter

with open('data.csv.txt','r') as f:
    csvreader = csv.DictReader(f)
            
    counta = Counter()
    for i in csvreader:
        c = i["LanguagesWorkedWith"].split(";")
        counta.update(c)
        
print(counta)
     
xvalues = []
yvalues = []

for x,y in counta.items():
    xvalues.append(x)
    yvalues.append(y)

print(xvalues)
print(yvalues)
    
plt.barh(xvalues,yvalues) 


#to change the size of chart
plt.rcParams['figure.figsize'] = [10, 7]
plt.show()
        
        
    
```
